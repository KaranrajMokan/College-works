<h1>Question-5</h1>

<p> &nbsp;&nbsp;&nbsp; The scatterplot of the two variables - carbohydrates and calories is given below.</p>

```python
carbo=df['carbo']
calories=df['calories']
fig,ax = plt.subplots()
ax.set_title("Scatterplot - Carbohydrates VS Calories")
ax.set_xlabel("Carbohydrates")
ax.set_ylabel("Calories")
ax.scatter(carbo,calories)
plt.show()

correlation = df['carbo'].corr(df['calories'])
print("The correlation is " + str(correlation))
```

According to the scatterplot, these two variables have a very weak linear relationship. Based on the correalation value between the two variables, they have a positive correlation because both the variables change in the same direction (+ve value). Since the correlation value is less than 0.3, it can be said that they have a very weak positive correlation.
